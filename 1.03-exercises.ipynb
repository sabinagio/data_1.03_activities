# Activity 1

```python
import pandas as pd
import numpy as np

# Load data into Jupyter notebook

df3 = pd.read_excel('file3.xlsx')
df4 = pd.read_excel('file4.xlsx')
```

## 1. Print the columns of the dataframes

```python
print(df3.columns)
print(df4.columns)
```

## 2. Change the name of the columns that were changed in class

```python
df3 = df3.rename(columns = {'CONTROLN': 'ID', 'HV1': 'Median_home_value', 'IC1': 'median_household_income'})
df4 = df4.rename(columns = {'CONTROLN': 'ID', 'HV1': 'Median_home_value', 'IC1': 'median_household_income'})
```

## 3. Change the column names to lowercase

```python
df3.columns = df3.columns.str.lower()
df4.columns = df4.columns.str.lower()
```

# Activity 2

## 1. Merge the previous 2 dataframes to the dataframe used in class

### 1.1. Read the dataframe from the class

```python

data = pd.read_csv('L03-files/file1.csv')
```

### 1.2. Check that the columns are in the same order in all data frames

```python
print(df3.columns)
print(df4.columns)
print(data.columns)
```

### 1.3. Change the data columns to match df3 & df4

```python
data.columns = df3.columns
data.head()
```

### 1.4. Merge the 3 dataframes

```python
final_data = pd.concat([data, df3, df4])
print(final_data.shape)
print(final_data.columns)
final_data.head()
```

## 2. Drop domain & dob columns

```python
final_data = final_data.drop(['domain','dob'], axis = 1)
print(final_data.columns)
```

## 3. Rearrange columns

```python
columns = final_data.columns
print(columns)

reordered_columns = ['id', 'state', 'gender', 'ic2', 'ic3', 'ic4', 'ic5',
 'median_home_value', 'median_household_income', 'hvp1', 
 'pobc1', 'pobc2', 'avggift', 'tcode', 'target_d']

final_data = final_data.reindex(columns = reordered_columns)
final_data.head()
```

## 4. Filter the rows for men that live in florida

### 4.1. Look at the unique values for state & gender

```python
print(final_data['state'].unique())
print(final_data['gender'].unique())
```

### 4.2. Select all the data for the state (FL) and gender (M, Male, male)

```python
florida_men = final_data[(final_data['state'] == 'FL') & (final_data['gender'].isin(['M', 'Male', 'male']))]
print(florida_men.shape)
print(florida_men.head())
```

### 4.3. Check that we've filtered the data

```python
print(florida_men['state'].unique())
print(florida_men['gender'].unique())
```

## 5. Filter the rows for female donors that have donated less than 100. Do not store the results in data.

### 5.1 Check the unique values in the _gender_ column

```python
print(final_data['gender'].unique())
female_list = ['F', 'female', 'Female', 'feamale']
```

### 5.2 Filter by gender and donation (_avggift_)

```python
female_donors = final_data[(final_data['gender'].isin(female_list)) & (final_data['avggift'] < 100)]
print(female_donors.head())
print(female_donors.shape)
```
